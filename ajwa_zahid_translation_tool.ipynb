```python
# Install Hugging Face Transformers and Gradio
!pip install transformers gradio -q

from transformers import pipeline
import gradio as gr

# Load a translation pipeline (English ↔ Urdu example)
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ur")

# Function for translation
def translate_text(text):
    result = translator(text, max_length=100)
    return result[0]['translation_text']

# Gradio Interface
iface = gr.Interface(
    fn=translate_text,
    inputs=gr.Textbox(label="Enter text in English"),
    outputs=gr.Textbox(label="Translation in Urdu"),
    title="AI Language Translation Tool",
    description="Translate text between English and Urdu"
)

iface.launch()
```

```python
# 📌 Step 1: Install libraries
!pip install transformers gradio -q

# 📌 Step 2: Import GPT-2
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import gradio as gr

# 📌 Step 3: Load GPT-2 model
model_name = "gpt2"  # try also "gpt2-medium" or "gpt2-large" for better quality
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# 📌 Step 4: Translation function using GPT-2
def gpt2_translate(text, target_lang):
    prompt = f"Translate this English sentence into {target_lang}:\n{text}\nTranslation:"
    result = generator(prompt, max_length=120, num_return_sequences=1, do_sample=True, temperature=0.7)
    return result[0]['generated_text'].split("Translation:")[-1].strip()

# 📌 Step 5: Gradio Interface
iface = gr.Interface(
    fn=gpt2_translate,
    inputs=[gr.Textbox(lines=3, placeholder="Enter English text here..."),
            gr.Dropdown(["French", "Spanish", "German"], label="Target Language")],
    outputs=gr.Textbox(label="Translated Text (GPT-2)"),
    title="🌍 GPT-2 Translator (Experimental)",
    description="Uses GPT-2 for translation. Results may be inaccurate, since GPT-2 is not fine-tuned for translation."
)

# 📌 Step 6: Launch app
iface.launch(share=True)
```
